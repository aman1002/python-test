##### 1. Create 4,16,....,52 week moving average(closing price) for each stock and index. This should happen through a function.

```python
def moving_average(iterable, win):
    iterb = iter(iterable)
    a = deque(itertools.islice(iterb, win-1))
    a.appendleft(0)
    S = sum(a)
    for item in iterb:
        S += item - a.popleft()
        a.append(item)
        yield S / float(win)

# Append null to first (win - 1) steps because we dont have points to calculate
# average till that point 
def window_average(iterable, win):
    result = [None for i in range(win-1)]
    result.extend(list(moving_average(iterable, win)))
    return result
```

```python
infy_4 = window_average(INFY['Close'], 20)
infy_16 = window_average(INFY['Close'], 80)
infy_28 = window_average(INFY['Close'], 140)
infy_40 = window_average(INFY['Close'], 200)
infy_52 = window_average(INFY['Close'], 260)
```

```python
tcs_4 = window_average(TCS['Close'], 20)
tcs_16 = window_average(TCS['Close'], 80)
tcs_28 = window_average(TCS['Close'], 140)
tcs_40 = window_average(TCS['Close'], 200)
tcs_52 = window_average(TCS['Close'], 260)
```

```python
nifty_4 = window_average(NIFTY_IT['Close'], 20)
nifty_16 = window_average(NIFTY_IT['Close'], 80)
nifty_28 = window_average(NIFTY_IT['Close'], 140)
nifty_40 = window_average(NIFTY_IT['Close'], 200)
nifty_52 = window_average(NIFTY_IT['Close'], 260)
```

##### 3.1 Volume shocks - If volume traded is 10% higher/lower than previous day - make a 0/1 boolean time series for shock, 0/1 dummy-coded time series for direction of shock.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
infy_vol = infy[['Volume']]
infy_vol['Previous'] = infy_vol.shift(1)
infy_vol.head()
```

```python
infy_vol = infy_vol[1:]
```

```python
tcs_vol = tcs[['Volume']]
tcs_vol['Previous'] = tcs_vol.shift(1)
tcs_vol = tcs_vol[1:]
```

```python
nifty_vol = nifty_it[['Volume']]
nifty_vol['Previous'] = nifty_vol.shift(1)
nifty_vol = nifty_vol[1:]
```

```python
infy_vol['Difference'] = infy_vol['Volume'] - infy_vol['Previous']
infy_vol['Volume Shock'] = abs(infy_vol['Difference']) > 0.1*infy_vol['Previous']
infy_vol.head()
```

```python
tcs_vol['Difference'] = tcs_vol['Volume'] - tcs_vol['Previous']
tcs_vol['Volume Shock'] = abs(tcs_vol['Difference']) > 0.1*tcs_vol['Previous']
tcs_vol.head()
```

```python
nifty_vol['Difference'] = nifty_vol['Volume'] - nifty_vol['Previous']
nifty_vol['Volume Shock'] = abs(nifty_vol['Difference']) > 0.1*nifty_vol['Previous']
nifty_vol.head()
```

```python
infy_vol['Direction_Shock'] = 'Nan'
tcs_vol['Direction_Shock'] = 'Nan'
nifty_vol['Direction_Shock'] = 'Nan'
```

```python
def direction(df):
    if df['Volume Shock'] == False:
        pass
    else:
        if df['Difference'] < 0:
            return '0'
        else:
            return '1'
```

```python
infy_vol['Direction_Shock'] = infy_vol.apply(direction, axis=1)
infy_vol.head()
```

```python
tcs_vol['Direction_Shock'] = tcs_vol.apply(direction, axis=1)
```

```python
nifty_vol['Direction_Shock'] = nifty_vol.apply(direction, axis=1)
```

```python
def volume(df):
    if df['Volume Shock'] == True:
        return '1'
    else:
        return '0'
```

```python
infy_vol['Volume Shock'] = infy_vol.apply(volume, axis=1)
infy_vol.head()
```

```python
tcs_vol['Volume Shock'] = tcs_vol.apply(volume, axis=1)
nifty_vol['Volume Shock'] = nifty_vol.apply(volume, axis=1)
```

```python
tcs_vol.head()
```

```python
nifty_vol.head()
```

##### 3.2 Price shocks - If closing price at T vs T+1 has a difference > 2%, then 0/1 boolean time series for shock, 0/1 dummy-coded time series for direction of shock.

```python
infy_close = INFY[['Prev Close', 'Close']]
tcs_close = TCS[['Prev Close', 'Close']]
```

```python
infy_close.head()
```

```python
nifty_close = nifty_it[['Close']]
nifty_close['Prev Close'] = nifty_close.shift(1)
```

```python
nifty_close.head()
```

```python
nifty_close = nifty_close[1:]
```

```python
infy_close['Difference'] = infy_close['Prev Close'] - infy_close['Close']
infy_close['Shock'] = abs(infy_close['Difference']) > 0.02*infy_close['Prev Close']
print(infy_close['Shock'].value_counts())
```

```python
tcs_close['Difference'] = tcs_close['Prev Close'] - tcs_close['Close']
tcs_close['Shock'] = abs(tcs_close['Difference']) > 0.02*tcs_close['Prev Close']
print(tcs_close['Shock'].value_counts())
```

```python
nifty_close['Difference'] = nifty_close['Prev Close'] - nifty_close['Close']
nifty_close['Shock'] = abs(nifty_close['Difference']) > 0.02*nifty_close['Prev Close']
print(nifty_close['Shock'].value_counts())
```

```python
def direction(df):
    if df['Shock'] == False:
        pass
    else:
        if df['Difference'] < 0:
            return '0'
        else:
            return '1'
```

```python
infy_close['Direction_Shock'] = infy_close.apply(direction, axis=1)
tcs_close['Direction_Shock'] = tcs_close.apply(direction, axis=1)
nifty_close['Direction_Shock'] = nifty_close.apply(direction, axis=1)
```

```python
def close(df):
    if df['Shock'] == True:
        return '1'
    else:
        return '0'
```

```python
infy_close['Shock'] = infy_close.apply(close, axis=1)
tcs_close['Shock'] = tcs_close.apply(close, axis=1)
nifty_close['Shock'] = nifty_close.apply(close, axis=1)
```

##### 3.4 Pricing shock without volume shock (pswvs) - based on points a & b - Make a 0/1 dummy time series.

```python
infy_df = pd.DataFrame()
```

```python
infy_df['Price Shock'] = infy_close['Shock']
infy_df['Volume Shock'] = infy_vol['Volume Shock']
infy_df.head()
```

```python
tcs_df = pd.DataFrame()
tcs_df['Price Shock'] = tcs_close['Shock']
tcs_df['Volume Shock'] = tcs_vol['Volume Shock']
```

```python
nifty_df = pd.DataFrame()
nifty_df['Price Shock'] = nifty_close['Shock']
nifty_df['Volume Shock'] = nifty_vol['Volume Shock']
```

```python
def shock(df):
    if df['Price Shock'] == '1' and df['Volume Shock'] == '0':
        return '1'
    else:
        return '0'
```

```python
infy_df['pswvs'] = infy_df.apply(shock, axis=1)
tcs_df['pswvs'] = tcs_df.apply(shock, axis=1)
nifty_df['pswvs'] = nifty_df.apply(shock, axis=1)
```
